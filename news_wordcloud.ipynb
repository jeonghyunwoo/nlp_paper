```
from konlpy.tag import Okt
```

```
okt = Okt()
```

```
tok = [w for w in texts if len(w) >1]
```

```
tok = [t for t in texts]
```

```
token = [okt.nouns(t) for t in tok]
```

```
token = [w for t in token for w in t if len(w) >1]
```

```
tok1 = pd.Series(token)
```

```
tok2 = tok1.value_counts()
```

```
tok3 = tok2.loc[tok2>1]
```

```
len(tok3)
```

```
tdic = {key : value for key, value in zip(tok3.index,tok3)}
```
