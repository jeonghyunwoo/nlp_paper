```python
import numpy as np
```

```python
periods = [720,-720] + ['Y'+str(y) for y in np.arange(2018,2012,-1)]+['Y2010-2012','Y2000-2009']
```

```python
pages = np.arange(1,1009,9)
```

```python
base_urls = [f"https://newslookup.com/results?p={y}&q=oil+price&dp=&mt=-1&mkt=0&mtx=0&mktx=0&s=&groupby=no&cat=-1&from=&fmt=&tp={x}&ovs=1" for x in periods for y in pages]
```

```python
%%time # 33min 47s
tit, pre, dats = [],[],[]
sz = len(base_urls)
for i,url in enumerate(base_urls):
    a = soup(requests.get(url).text, 'lxml')
    titl = [t.get_text() for t in a.find_all('a','title')]
    pres = [p['href'].split('/')[2] for p in a.find_all('a','title')]
    dat = [d.get_text() for d in a.find_all('span','stime')]
    tit.append(titl)
    pre.append(pres)
    dats.append(dat)
    print(i,'/',sz,'done')
```

```python
import pandas as pd
```

```python
tit1 = [tt for t in tit for tt in t]
```

```python
pre1 = [pp for p in pre for pp in p]
```

```python
dats1 = [dd for d in dats for dd in d]
```

```python
tsd = pd.DataFrame({'tit':tit1,'pres':pre1,'dat':dats1})
```
